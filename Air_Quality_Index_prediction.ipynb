# For time series data we do forecasting and here we use fb prophet model. https://facebook.github.io/prophet/docs/quick_start.html

```python
date_info = pd.to_datetime(air_quality_data['Date'], dayfirst=True)
print(date_info)
```

```python
date_info = date_info.dt.strftime('%Y-%mz-%d')
print(date_info)
```

lambda is used as a single line function.


```python
time_info = air_quality_data['Time']
print(time_info)
```

```python
time_info = time_info.apply(lambda x: x.replace('.' , ':'))
print(time_info)
```

```python
print(type(time_info))
print(type(date_info))
```

```python
date_time = pd.concat([date_info, time_info], axis=1)
print(date_time)
```

```python
date_time.shape
```

```python
date_time['ds'] = date_time['Date'].astype(str)+' '+ date_time['Time'].astype(str)
```

```python
date_time.info()
```

```python
data = pd.DataFrame()
```

```python
data['ds'] = pd.to_datetime(date_time['Date'].astype(str) + ' ' + date_time['Time'].astype(str))
print(data)
```

```python
data['y'] = air_quality_data['RH']
data.head()
```

```python
pip install prophet
```

```python
from prophet import Prophet
```

```python
model = Prophet()
model.fit(data)
```

```python
future = model.make_future_dataframe(periods=365)
future.tail()
```

```python
forecast = model.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

'yhat' is indicating the middle value.

```python
fig1 = model.plot(forecast)
```

```python
fig2 = model.plot_components(forecast)
```

```python
from prophet.plot import plot_plotly, plot_components_plotly

plot_plotly(model, forecast)
```

```python
plot_components_plotly(model, forecast)
```
